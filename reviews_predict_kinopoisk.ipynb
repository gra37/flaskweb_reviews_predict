### Предварительная обработка данных (Preprocessing of Data)

```python
import re
import string
#import fasttext
import os

from sklearn.feature_extraction.text import TfidfTransformer

from nltk.corpus import stopwords

from nltk.tokenize import word_tokenize   # делит кадждое слово на токены '<', '3',
from nltk.tokenize import TweetTokenizer  #при токениации некторые слова не раделяет например '<3'
from nltk.tokenize import sent_tokenize

from nltk.stem import WordNetLemmatizer
from nltk.stem.snowball import SnowballStemmer

from sklearn.model_selection import train_test_split



tokenizer = TweetTokenizer()
#stemmer = SnowballStemmer("english")
#stop_words = stopwords.words('english') # определим стоп слова
#stop_words.remove('not') # оставим слово "не" - оно нам может пригодиться

stemmer = SnowballStemmer("russian")
stop_words = stopwords.words('russian')
stop_words.remove('не') # оставим слово "не" - оно нам может пригодиться

lemmatizer = WordNetLemmatizer()           #  в NLTK нет русского лемматизатора, поэтому будем использовать стемминг,      
                                           #суть которого заключается в отбрасывании суффиксов и приставок у слов 

punct_dct = {ord(p): None for p in string.punctuation}
```

```python
def data_preprocessing(dialogue):
      
    # Очистка данных
    dialogue = re.sub(re.compile('<.*?>'), '', dialogue) #удалим html tags
    #dialogue =  re.sub('[^a-zA-Zа-яА-Я1-9]+', ' ', dialogue) #оставим только слова в тексте
    dialogue =  re.sub('[^а-яА-Я1-9]+', ' ', dialogue)
    dialogue =  re.sub(' +', ' ', dialogue)
    
    dialogue =  re.sub('((www\.[^\s]+)|(https?://[^\s]+))', 'URL', dialogue)
    
    dialogue = dialogue.translate(punct_dct)
    
    # Приведем все к нижнему регистру
    dialogue = dialogue.lower().replace("ё", "е").replace("\t", " ")
    dialogue = dialogue.replace("\n", " ").replace("\r", " ")

    # Токенизация (выбрать что то одно)
    #tokens = nltk.word_tokenize(dialogue) # конвертируем диалоги в токены (по словам)
    tokens = tokenizer.tokenize(dialogue)

    # Удалим стоп-слова
    tokens = [word for word in tokens if word not in stop_words] #Удалим стоп-слова из диалога

    # lemmatization
    #words = [lemmatizer.lemmatize(tokens) for word in tokens] # Приведем слова к их основанию (только для АНГЛ слов)
    words = [stemmer.stem(token) for token in tokens]
    
    # join words in preprocessed dialogue
    dialogue = ' '.join(words).strip()

    return dialogue
```

```python
df['обработанный отзыв'] = df['отзыв'].apply(lambda dialogue: data_preprocessing(dialogue))
df
```

## Splitting data(70-30): Train | Test

```python
from sklearn.model_selection import train_test_split

x = df['обработанный отзыв'].values
y = df['labels'].values
# data.drop(['label'], axis=1, inplace=True)

x_t, x_control, y_t, y_control = train_test_split(x, y, test_size=0.1)#, stratify=y)
x_train, x_test, y_train, y_test = train_test_split(x_t, y_t, test_size=0.3)#, stratify=y)

print("Train data:",  x_train.shape, y_train.shape)
print("Test data:",  x_test.shape, y_test.shape)
print("Control data:",  x_control.shape, y_control.shape)
```

## Vectorizing Text Data
#### TFIDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(sublinear_tf=True, min_df=0, norm='l2', encoding='latin-1')

x_train_dialogue_tfidf = vectorizer.fit_transform(x_train)
x_test_dialogue_tfidf = vectorizer.transform(x_test)
x_control_dialogue_tfidf = vectorizer.transform(x_control)

print('x_train_dialogue_tfidf shape: ', x_train_dialogue_tfidf.shape)
print('x_test_dialogue_tfidf shape: ', x_test_dialogue_tfidf.shape)
```

## Building Models

```python
import warnings
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
from sklearn import metrics
from sklearn.metrics import f1_score

warnings.filterwarnings('ignore')

from sklearn.linear_model import LogisticRegression

clf = LogisticRegression(penalty='l2')
clf.fit(x_train_dialogue_tfidf, y_train)

y_test_pred = clf.predict(x_test_dialogue_tfidf)

    
print('Test Accuracy: ', accuracy_score(y_test, y_test_pred))
print('Test F1_score: ', f1_score(y_test, y_test_pred, average=None)[0])
```

### RandomForest

```python
model = RandomForestClassifier(n_estimators=100)
model.fit(x_train_dialogue_tfidf, y_train)
```
